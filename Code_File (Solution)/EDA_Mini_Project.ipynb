### Q5. Count the number of data values that fall within two standard deviations of the mean. Compare this with the answer from Chebyshev’s Theorem.

#### Chebyshev's Rule
For any data set, regardless of distribution, the proportion of values that fall within \( k \) standard deviations of the mean is at least:
$$ 1 - \frac{1}{k^2} $$

where \( k \geq 1 \).

#### Empirical Rule (68-95-99.7 Rule) for Normal Distribution
For a normally distributed data set:
- Approximately 68% of the data values fall within one standard deviation of the mean:
  $$ [\mu - \sigma, \mu + \sigma] $$
- Approximately 95% of the data values fall within two standard deviations of the mean:
  $$ [\mu - 2\sigma, \mu + 2\sigma] $$
- Approximately 99.7% of the data values fall within three standard deviations of the mean:
  $$ [\mu - 3\sigma, \mu + 3\sigma] $$

where:
- \( \mu \) is the mean,
- \( \sigma \) is the standard deviation.


```python
# data points that falls in two standard deviation for normal distribution
# Calculate lower and upper limits for two standard deviations
lower_limit = b - 2 * std
upper_limit = b + 2 * std

# Count the number of data values within two standard deviations
data_within_2std = ages[(ages >= lower_limit) & (ages <= upper_limit)]
count_within_2std = len(data_within_2std)

print("Number of data points within two standard deviations of the mean:", count_within_2std)
```

```python
# data points that falls in two standard deviation for chebyshev rule

# Calculate Chebyshev's lower bound
chebyshev_prop = 1 - 1 / 2**2  # k = 2
chebyshev_count_lower = chebyshev_prop * len(ages)
print(f'Number of data points within two standard deviations of the mean : {round(chebyshev_count_lower)}')
```

```python

from scipy.stats import norm

# Calculate mean and standard deviation
mean_age = np.mean(ages)
std_age = np.std(ages)

# Calculate lower and upper limits for two standard deviations
lower_limit = mean_age - 2 * std_age
upper_limit = mean_age + 2 * std_age

# Count the number of data values within two standard deviations
data_within_2std = ages[(ages >= lower_limit) & (ages <= upper_limit)]
count_within_2std = len(data_within_2std)

print("Number of data values within two standard deviations of the mean:", count_within_2std)

# Plotting the normal distribution and highlighting the area within two standard deviations
plt.figure(figsize=(7, 5))
x = np.linspace(mean_age - 4*std_age, mean_age + 4*std_age, 1000)
p = norm.pdf(x, mean_age, std_age)
plt.plot(x, p, 'k', linewidth=2)
plt.axvline(lower_limit, color='red', linestyle='--', label='Lower Limit (2 std)')
plt.axvline(upper_limit, color='green', linestyle='--', label='Upper Limit (2 std)')
plt.fill_between(x, 0, p, where=(x >= lower_limit) & (x <= upper_limit), color='yellow', alpha=0.3)
plt.xlabel('Age')
plt.ylabel('Probability Density')
plt.title('Normal Distribution with Two Standard Deviations')
plt.legend()

# Annotating the lower and upper limits
plt.text(lower_limit, norm.pdf(lower_limit, mean_age, std_age)/2, f'{lower_limit:.2f}', color='red', ha='center', va='center', backgroundcolor='w')
plt.text(upper_limit, norm.pdf(upper_limit, mean_age, std_age)/2, f'{upper_limit:.2f}', color='green', ha='center', va='center', backgroundcolor='w')

plt.show()

# Chebyshev's rule calculations
chebyshev_prop = 1 - 1 / 2**2  # k = 2
chebyshev_count_lower = chebyshev_prop * len(ages)

print("Chebyshev's rule lower bound count:", chebyshev_count_lower)
print("Proportion of data within two standard deviations according to Chebyshev's rule:", chebyshev_prop)
print("Actual proportion of data within two standard deviations:", count_within_2std / len(ages))
```

### Q6. Find the three quartiles and the interquartile range (IQR).

### Quartiles and Interquartile Range (IQR)

#### First Quartile (Q1)
The first quartile, \(Q1\), is the median of the lower half of the data (25th percentile).
$$ Q1 = \left(\frac{n+1}{4}\right)^{th} \text{ value} $$

#### Second Quartile (Q2)
The second quartile, \(Q2\), is the median of the data (50th percentile).
$$ Q2 = \left(\frac{n+1}{2}\right)^{th} \text{ value} $$

#### Third Quartile (Q3)
The third quartile, \(Q3\), is the median of the upper half of the data (75th percentile).
$$ Q3 = \left(\frac{3(n+1)}{4}\right)^{th} \text{ value} $$

#### Interquartile Range (IQR)
The interquartile range (IQR) is the range of the middle 50% of the data.
$$ IQR = Q3 - Q1 $$


```python
Q1 = np.quantile(ages, 0.25)
Q2 = np.quantile(ages, 0.50)
Q3 = np.quantile(ages, 0.75)

IQR = Q3 - Q1

print(f"First Quartile (Q1): {Q1}")
print(f"Second Quartile (Q2 or Median): {Q2}")
print(f"Third Quartile (Q3): {Q3}")
print(f"Interquartile Range (IQR): {IQR}")
```

### Q14. Treat this dataset as a binomial distribution where p is the probability that a person becomes a CEO above 50 years age. What is the probability that out of a random sample of 10 CEOs of Fortune 500 companies exactly 6 are above 50 years of age?


The binomial probability mass function (PMF) is given by:

$$ P(X = k) = \binom{n}{k} \cdot p^k \cdot (1 - p)^{n - k} $$

Where:
- \( n \) is the number of trials (CEOs sampled),
- \( k \) is the number of successes (CEOs above 50 years),
- \( p \) is the probability of success (probability that a CEO is above 50 years of age),
- \( \binom{n}{k} \) is the binomial coefficient, representing the number of ways to choose \( k \) successes from \( n \) trials.


```python
import matplotlib.pyplot as plt
from scipy.stats import binom

# Number of CEOs sampled
n = 10

# Number of CEOs above 50 years of age
k = 6

# Probability that a person becomes a CEO above 50 years of age
p = 0.5

# Calculate the probability using the binomial probability mass function
probability = binom.pmf(k, n, p)

print("Probability:", probability)

# Plotting the probability in a pie chart
labels = ['Success (6 CEOs above 50)', 'Failure (4 CEOs below 50)']
sizes = [probability, 1 - probability]
colors = ['#FF6347', '#87CEEB']  # Using Coral and Sky Blue colors
explode = (0.1, 0)  # Explode the "Success" slice

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, shadow=True, explode=explode)
plt.title('Binomial Probability Distribution for CEOs above 50', fontsize=16, fontweight='bold', color='navy')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.legend(title="Legend", loc="best", fontsize=12)

# Adding a background color
plt.gca().set_facecolor('#f7f7f7')

plt.show()
```

### Q15. A study claims that 25% of all Fortune 500 companies becoming a CEO are above 60 years of age. Using the Normal approximation of a Binomial distribution, find the probability that in a random sample of 300 Fortune 500 companies exactly 75 CEOs will be above 50 years of age. [ Note that the normal distribution can be used to approximate a binomial distribution if np>=5 and nq>=5 with the following correction for continuity P(X=z) = P(z-0.5 < X < z+0.5 ) ]

To approximate the probability using the Normal distribution for a Binomial distribution:

1. **Probability Mass Function (Binomial):**

   The binomial probability mass function (PMF) is:
   
   $$ P(X = k) = \binom{n}{k} \cdot p^k \cdot (1 - p)^{n - k} $$
   
   Where:
   - \( n \) is the number of trials (number of Fortune 500 companies sampled),
   - \( k \) is the number of successes (CEOs above 50 years of age),
   - \( p \) is the probability of success (probability that a CEO is above 60 years of age),
   - \( \binom{n}{k} \) is the binomial coefficient, representing the number of ways to choose \( k \) successes from \( n \) trials.

2. **Normal Approximation:**

   If \( np \geq 5 \) and \( n(1-p) \geq 5 \), the Binomial distribution can be approximated by a Normal distribution with mean \( \mu = np \) and variance \( \sigma^2 = np(1-p) \).
   
   The continuity correction is applied as follows:
   
   $$ P(X = k) \approx P\left( k - 0.5 < X < k + 0.5 \right) $$
   
   Where \( X \) follows a Normal distribution \( \mathcal{N}(\mu, \sigma^2) \).
   
3. **Given Example:**

   Given that 25% of CEOs are above 60 years of age:
   - \( n = 300 \) (number of Fortune 500 companies sampled),
   - \( k = 75 \) (number of CEOs above 50 years of age),
   - \( p = 0.25 \) (probability that a CEO is above 60 years of age).

   Therefore, the probability using Normal approximation is:
   
   $$ P(X = 75) \approx P\left( 74.5 < X < 75.5 \right) $$
   
   Where \( X \) is approximately \( \mathcal{N}(np, np(1-p)) \).


```python
import numpy as np
from scipy.stats import norm

n = 300   # number of trials (Fortune 500 companies sampled)
p = 0.25  # probability of a CEO being above 60 years of age
k = 75    # number of CEOs above 50 years of age we are interested in

# Calculate mean and standard deviation of the binomial distribution
mu = n * p
sigma = np.sqrt(n * p * (1 - p))

# Calculate the probability using normal approximation with continuity correction
z_lower = (k - 0.5 - mu) / sigma
z_upper = (k + 0.5 - mu) / sigma
probability = norm.cdf(z_upper) - norm.cdf(z_lower)

print(f"Probability that exactly 75 CEOs will be above 50 years of age: {probability:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

n = 300
p = 0.25
k = 75

# Calculate mean and standard deviation of the binomial distribution
mu = n * p
sigma = np.sqrt(n * p * (1 - p))

# Calculate the probability using normal approximation with continuity correction
z_lower = (k - 0.5 - mu) / sigma
z_upper = (k + 0.5 - mu) / sigma
probability = norm.cdf(z_upper) - norm.cdf(z_lower)

print("Probability:", probability)

# Plotting the probability in a pie chart with advanced decorations
labels = ['Success (75 out of 300)', 'Failure (225 out of 300)']
sizes = [probability, 1 - probability]
colors = ['#66c2a5', '#fc8d62']  # Custom colors

plt.figure(figsize=(6, 6))

# Explode a slice to emphasize it
explode = (0.1, 0)  # Explode the first slice (Success)

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, explode=explode, shadow=True)

# Add a title with advanced styling
plt.title('Normal Approximation with Continuity Correction', fontsize=16, fontweight='bold', color='blue', loc='center')

# Add a legend with custom font size and border
plt.legend(labels, loc='best', fontsize=12, fancybox=True, framealpha=0.7, edgecolor='black')

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Show the plot
plt.show()
```

### Q16. Compute a 95% Confidence Interval for the true mean age of the population of CEOs for the given dataset using appropriate distribution.( State reasons as to why did you use a z or t distribution)

The confidence interval (CI) for the population mean (\( \mu \)) using the t-distribution is given by:

$$ \bar{x} \pm t_{\alpha/2, n-1} \cdot \frac{s}{\sqrt{n}} $$

Where:
- \( \bar{x} \) is the sample mean,
- \( t_{\alpha/2, n-1} \) is the critical value from the t-distribution with \( n-1 \) degrees of freedom and significance level \( \alpha/2 \),
- \( s \) is the sample standard deviation,
- \( n \) is the sample size.

The critical value \( t_{\alpha/2, n-1} \) can be found using statistical tables or computational tools based on the desired confidence level and degrees of freedom.


```python
import numpy as np
from scipy.stats import t

# Compute sample mean and standard deviation
sample_mean = np.mean(ages)
sample_std = np.std(ages, ddof=1)  # ddof=1 for sample standard deviation

# Calculate the standard error of the mean
n = len(ages)
standard_error = sample_std / np.sqrt(n)

# Degrees of freedom
df = n - 1

# Calculate the margin of error using the T-distribution
alpha = 0.05  # 1 - confidence level
t_critical = t.ppf(1 - alpha/2, df)
margin_of_error = t_critical * standard_error

# Compute the confidence interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print("95% Confidence Interval:", np.round(confidence_interval,2))
```

### Q17. A data scientist wants to estimate with 95% confidence the proportion of CEOs of Fortune 500 companies are above 60 years in the population. Another recent study showed that 25% of CEOs interviewed were above 60. The data scientist wants to be accurate within 2% of the true proportion. Find the minimum sample size necessary.

The formula for minimum sample size (n) required to estimate the population mean (\( \mu \)) with a specified margin of error (E) at a given confidence level (1 - \( \alpha \)) is:

$$ n = \left( \frac{z_{\alpha/2} \cdot \sigma}{E} \right)^2 $$

Where:
- \( z_{\alpha/2} \) is the critical value from the standard normal distribution corresponding to the desired confidence level (e.g., 1.96 for 95% confidence),
- \( \sigma \) is the population standard deviation (if known),
- \( E \) is the desired margin of error (maximum allowable error in estimating \( \mu \)).


The formula for minimum sample size (n) required to estimate a population proportion (p) with a specified margin of error (E) at a given confidence level (1 - \( \alpha \)) is:

$$ n = \left( \frac{z_{\alpha/2}^2 \cdot p \cdot (1 - p)}{E^2} \right) $$

Where:
- \( z_{\alpha/2} \) is the critical value from the standard normal distribution corresponding to the desired confidence level (e.g., 1.96 for 95% confidence),
- \( p \) is the estimated population proportion,
- \( E \) is the desired margin of error, which is the maximum allowable error in estimating \( p \).


```python
import math

# Given values
z = 1.96  # Z-score for 95% confidence level
p = 0.25  # Estimated proportion
E = 0.02  # Margin of error

# Calculate sample size
n = (z**2 * p * (1 - p)) / E**2

# Round up to the nearest whole number
n = math.ceil(n)

print("Minimum sample size necessary:", n)
```

### Q19. A researcher claims that currently 25% of all CEOS are above 60 years .Test his claim with an alpha = 0.05 if out of a random sample of 30 CEOs only 10 are above 60 years.


The z-test formula for testing a population mean (\( \mu \)) against a null hypothesis value (\( \mu_0 \)) is:

$$ z = \frac{\bar{x} - \mu_0}{\frac{\sigma}{\sqrt{n}}} $$

Where:
- \( \bar{x} \) is the sample mean,
- \( \mu_0 \) is the hypothesized population mean under the null hypothesis,
- \( \sigma \) is the population standard deviation (if known),
- \( n \) is the sample size.

The z-score \( z \) indicates how many standard deviations the sample mean is from the hypothesized population mean (\( \mu_0 \)).


The z-test formula for testing a population proportion (p) against a null hypothesis value (p0) is:

$$ z = \frac{\hat{p} - p_0}{\sqrt{\frac{p_0 \cdot (1 - p_0)}{n}}} $$

Where:
- \( \hat{p} \) is the sample proportion,
- \( p_0 \) is the hypothesized population proportion under the null hypothesis,
- \( n \) is the sample size.

The z-score \( z \) indicates how many standard deviations the sample proportion is from the hypothesized population proportion (\( p_0 \)).


```python
import math
from statsmodels.stats.proportion import proportions_ztest

H0 = 'The proportion of CEOs above 60 years old is 25%'
H1 = 'The proportion of CEOs above 60 years old is not 25%'

# Given values
p_proportion = 0.25  # Hypothesized population proportion
n = 30  # Sample size
x = 10  # Number of CEOs above 60 years in the sample
alpha = 0.05  # Significance level

# Calculate the sample proportion
p_sample = x / n

# Perform the z-test for proportions
test_stat, p_value = proportions_ztest(count=x, nobs=n, value=p_proportion, alternative='two-sided')

# Make a decision
if p_value < alpha:
    print(f"Reject the null hypothesis: {H1}")
else:
    print(f"Fail to reject the null hypothesis: {H0}")

print(f"p-value: {p_value}")
print(f"Test statistic: {test_stat}")
```

## c. What statistical test will you use?

Two-Sample t-Test: Used to compare the means of two independent groups.
    
Assumptions:
1. Independence: The samples from the two groups must be independent.
2. Normality: Data should be approximately normally distributed within each group.
3. Equal Variances: Ideally, the variances should be equal. If not, use Welch's t-test.

## d.What are the test results? (Assume s1 = 8.8 and s2= 7.8.)

The formula for the two-sample t-test compares the means of two independent samples to determine if they are significantly different:

$$ t = \frac{\bar{X}_1 - \bar{X}_2}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}} $$

Where:
- \( \bar{X}_1 \) and \( \bar{X}_2 \) are the sample means of the two samples,
- \( s_1^2 \) and \( s_2^2 \) are the sample variances of the two samples,
- \( n_1 \) and \( n_2 \) are the sample sizes of the two samples.

The degrees of freedom (df) for the t-test is given by:

$$ df = \frac{\left( \frac{s_1^2}{n_1} + \frac{s_2^2}{n_2} \right)^2}{\frac{\left( \frac{s_1^2}{n_1} \right)^2}{n_1 - 1} + \frac{\left( \frac{s_2^2}{n_2} \right)^2}{n_2 - 1}} $$

The t-statistic \( t \) measures the difference between the means relative to the variability within the samples, indicating whether the observed difference is statistically significant.


```python
import numpy as np
from scipy import stats

# Given data
uk_ceo_ages = np.array([47, 49, 73, 50, 65, 70, 49, 47, 40, 43, 46, 35, 38, 40, 47, 39, 49, 37, 37, 36, 40, 37, 31, 48, 48, 45, 52, 38, 38, 36, 44, 40, 48, 45, 45, 36, 39, 44, 52, 47])
usa_ceo_ages = np.array([47, 57, 52, 47, 48, 56, 56, 52, 50, 40, 46, 43, 44, 51, 36, 42, 49, 49, 40, 43, 39, 39, 22, 41, 45, 46, 39, 32, 36, 32, 32, 32, 37, 33, 44, 49, 44, 44, 49, 32])

# Given standard deviations
s1 = 8.8
s2 = 7.8

# Sample sizes
n1 = len(uk_ceo_ages)
n2 = len(usa_ceo_ages)

# Calculate means
mean_uk = np.mean(uk_ceo_ages)
mean_usa = np.mean(usa_ceo_ages)

# Calculate standard error of the difference between means
se = np.sqrt((s1**2 / n1) + (s2**2 / n2))

# Calculate t-statistic
t_statistic = (mean_uk - mean_usa) / se

# Calculate degrees of freedom for unequal variances
df = ((s1**2 / n1) + (s2**2 / n2))**2 / ((s1**2 / n1)**2 / (n1 - 1) + (s2**2 / n2)**2 / (n2 - 1))

# Calculate p-value
p_value = stats.t.sf(np.abs(t_statistic), df) * 2

# Print results
print(f"Mean age of UK CEOs: {mean_uk}")
print(f"Mean age of USA CEOs: {mean_usa}")
print(f"T-statistic: {round(t_statistic,3)}")
print(f"Degrees of freedom: {round(df,3)}")
print(f"P-value: {round(p_value,3)}")
```

### 12. With the help of a standard scaler, normalize the respective column distribution.


The formula for standard scaling (z-score normalization) of a dataset is:

$$ z = \frac{x - \mu}{\sigma} $$

Where:
- \( z \) is the standardized value,
- \( x \) is the original value,
- \( \mu \) is the mean of the dataset,
- \( \sigma \) is the standard deviation of the dataset.

Standard scaling transforms the data such that the resulting distribution has a mean of 0 and a standard deviation of 1.

### 18. With the help of hypothesis testing check whether fraudulent transactions of higher value than normal transactions?

```python
from scipy.stats import shapiro, levene, mannwhitneyu
from scipy.stats import ttest_ind

H0 = 'There is no Higher values of fraudulent transactions compared to non-fraudulent transactions'
H1 = 'Fraudulent transactions involve higher values than normal transactions'

fraudulent_transactions = df[df['Class'] == 1]['Amount']
non_fraudulent_transactions = df[df['Class'] == 0]['Amount']

# Normality check for x
test_stats, p_value_x = shapiro(fraudulent_transactions)
print('p_value for x :', p_value_x)
if p_value_x <= 0.05:
    print('Data is Not Normal')
else:
    print('Data is Normal')

print()

# Normality check for y
test_stats, p_value_y = shapiro(non_fraudulent_transactions)
print('p_value for y :', p_value_y)
if p_value_y <= 0.05:
    print('Data is Not Normal')
else:
    print('Data is Normal')

print()

# Variance check
test_stats, p_value_var = levene(fraudulent_transactions, non_fraudulent_transactions)
print('variances p_value :', p_value_var)
if p_value_var <= 0.05:
    print('Variances are Not Equal')
else:
    print('Variances are Equal')
    
print()

# Mann-Whitney U test (non-parametric, two groups)
u_stat, p_value_mw = mannwhitneyu(fraudulent_transactions, non_fraudulent_transactions,alternative = 'greater')
print('whitney score :', round(u_stat, 2))
print('P_value for whitney :', p_value_mw)

alpha = 0.05
if p_value_mw < alpha:
    print(f'Reject Null Hypothesis\n        ------------> {H1}')
else:
    print(f'Failed to Reject Null Hypothesis : {H0}')
    
print()

# Step 5: Draw the conclusion
conclusion = f"U statistic: {u_stat}, p-value: {p_value_mw}"
print(conclusion)
```

### 19. Perform ANOVA test for Statistical feature selection.

```python
import pandas as pd
from scipy.stats import f_oneway, shapiro, levene

# Assuming df is your DataFrame containing the data

# Separate the features and target variable
features = df.drop(columns = ['Class'])
target = df['Class']

# Separate the data into two groups based on the target variable
fraudulent_data = df[df['Class'] == 1]
normal_data = df[df['Class'] == 0]

# Perform the checks and tests for each feature
results = []

for feature in features.columns:
    fraud_values = fraudulent_data[feature]
    normal_values = normal_data[feature]

    # Normality check using Shapiro-Wilk test
    shapiro_fraud = shapiro(fraud_values)
    shapiro_normal = shapiro(normal_values)

    # Variance check using Levene's test
    levene_test = levene(fraud_values, normal_values)

    # ANOVA test
    f_stat, p_value = f_oneway(fraud_values, normal_values)

    # Determine significance based on ANOVA p-value
    if p_value < 0.05:
        significance = 'Yes'
    else:
        significance = 'No'

    results.append({
        'Feature': feature,
        'Shapiro P-value (Fraud)': shapiro_fraud.pvalue,
        'Shapiro P-value (Non-Fraud)': shapiro_normal.pvalue,
        'Levene P-value': levene_test.pvalue,
        'ANOVA P-value': p_value,
        'Significance': significance
    })

# Convert results to DataFrame
results_df = pd.DataFrame(results)

# Print the results
results_df
```
